# Customer Support Prompt Engineering Lab

In this notebook, we will test **Zero-Shot**, **Few-Shot**, and **Chain-of-Thought (CoT)** prompting out using the Groq API.

```python
!pip install groq python-dotenv
import os
from groq import Groq
from dotenv import load_dotenv

load_dotenv() # Load from .env file
client = Groq(api_key=os.environ.get("GROQ_API_KEY", "YOUR_API_KEY"))

def analyze_prompt(prompt):
    response = client.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model="llama-3.3-70b-versatile",
        temperature=0.5,
    )
    print(response.choices[0].message.content)
```

### 1. Zero-Shot Prompting

```python
zero_shot = """You are a polite customer support bot.
Answer the query: My mug arrived shattered."""
analyze_prompt(zero_shot)
```

### 2. Few-Shot Prompting

```python
few_shot = """You are a helpful customer support bot.

Customer: Where is my order? 
Support: I can help! Please provide your order number.

Customer: The app is slow!
Support: I apologize for the issue. Could you share your device details?

Now answer:
Customer: My mug arrived shattered.
Support:"""
analyze_prompt(few_shot)
```

### 3. Chain-of-Thought Prompting

```python
cot_prompt = """You are a support bot. Think step-by-step before answering.
1. Identify intent
2. Determine needed info
3. Formulate polite answer

Use format:
<thinking>
Step 1: ...
</thinking>
Response: ...

Customer: My mug arrived shattered."""
analyze_prompt(cot_prompt)
```
